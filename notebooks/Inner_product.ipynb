# Understanding Vectors, Lengths, and Angles

In this snippet, we explore the concept of **vectors** in a 2D space, how to calculate their **lengths**, and the **angle** between them using **inner products (dot products)**. Here’s a breakdown of what we learn:

#### Vectors
A **vector** is a list of numbers where the order matters, it corresponds to a mathematical object that has both a **magnitude** (length) and a **direction**. In this example, we work with two vectors:

-  $$\mathbf{v_1} = [1, 0] $$ (a vector pointing along the x-axis)
-  $$ \mathbf{v_2} = [-1, 0] $$ (a vector pointing in the opposite direction along the x-axis)


````python
import numpy as np
import matplotlib.pyplot as plt

# Define two 2D vectors
v1 = np.array([1, 0])  # Example vector 1
v2 = np.array([-1, 0])  # Example vector 2

# Plot the vectors
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

# Draw vectors
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')

# Add grid and labels
ax.grid(True, linestyle='--', linewidth=0.5)
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('2D Vectors and Inner Product')
ax.legend()

plt.show()
````

#### Length of Vectors (Magnitude)
The **length** (or **magnitude**) of a vector $ \mathbf{v} = [v_1, v_2] $ in a 2D space is given by the **Euclidean norm**:

$$
\text{Length of } \mathbf{v} = \| \mathbf{v} \| = \sqrt{v_1^2 + v_2^2}
$$

For example:
- The length of vector $ \mathbf{v_1}$  is $$ \| \mathbf{v_1} \| = \sqrt{1^2 + 0^2} = 1 $$.
- The length of vector $ \mathbf{v_2} $ is $$ \| \mathbf{v_2} \| = \sqrt{(-1)^2 + 0^2} = 1 $$.


````python
import numpy as np

# Define two 2D vectors
v1 = np.array([1, -1])  # Example vector 1
v2 = 1*np.array([1, 1])  # Example vector 2

# Compute magnitudes (lengths) of the vectors
length_v1 = np.linalg.norm(v1)
length_v2 = np.linalg.norm(v2)

# Print lengths
print(f"Vector 1: {v1}, Length: {length_v1:.2f}")
print(f"Vector 2: {v2}, Length: {length_v2:.2f}")
````

#### Inner Product (Dot Product)
The **inner product** (or **dot product**) of two vectors $ \mathbf{v_1} = [v_{1x}, v_{1y}] $ and $ \mathbf{v_2} = [v_{2x}, v_{2y}]$ is calculated as:

$$
\mathbf{v_1} \cdot \mathbf{v_2} = (v_{1x} \times v_{2x}) + (v_{1y} \times v_{2y})
$$

For our vectors:
$$
\mathbf{v_1} \cdot \mathbf{v_2} = (1 \times -1) + (0 \times 0) = -1
$$

This inner product tells us about the **relationship** between the two vectors. If the inner product is 0, the vectors are **perpendicular**. If the inner product is positive, the vectors point in a similar direction, and if negative, they point in opposite directions.

````python
# Compute inner product (dot product)
inner_product = np.dot(v1, v2)

# Print inner product result
print(f"Inner product is: ({v1[0]} * {v2[0]}) + ({v1[1]} * {v2[1]}) = {v1[0] * v2[0]} + {v1[1] * v2[1]} = {inner_product}")
````


#### Angle Between Vectors
The **cosine** of the angle \( \theta \) between two vectors is given by the formula:

$$
\cos(\theta) = \frac{\mathbf{v_1} \cdot \mathbf{v_2}}{\|\mathbf{v_1}\| \|\mathbf{v_2}\|}
$$

For our vectors:
$$
\cos(\theta) = \frac{-1}{1 \times 1} = -1
$$


This tells us that the two vectors are **antiparallel**, meaning they point in opposite directions.

````python
# Compute the cosine of the angle between the vectors
cos_theta = inner_product / (length_v1 * length_v2)
theta = np.arccos(cos_theta)  # Compute theta in radians
theta_degrees = np.degrees(theta)  # Convert to degrees

# Print angle in degrees
print(f"Cosine of the angle: {cos_theta:.2f}")
print(f"Angle θ: {theta_degrees:.2f} degrees")
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define two 2D vectors
v1 = np.array([1, -1])  # Example vector 1
v2 = 5*np.array([1, 1])  # Example vector 2

# Compute inner product (dot product)
inner_product = np.dot(v1, v2)

# Compute magnitudes (lengths) of the vectors
length_v1 = np.linalg.norm(v1)
length_v2 = np.linalg.norm(v2)

# Compute the cosine of the angle between the vectors
cos_theta = inner_product / (length_v1 * length_v2)
theta = np.arccos(inner_product / (length_v1 * length_v2))  # Compute theta in radians
theta_degrees = np.degrees(theta)  # Convert to degrees

# Print results
print(f"Vector 1: {v1}, Length: {length_v1:.2f}")
print(f"Vector 2: {v2}, Length: {length_v2:.2f}")
print(f"Inner product is: ({v1[0]} * {v2[0]}) + ({v1[1]} * {v2[1]}) = {v1[0] * v2[0]} + {v1[1] * v2[1]} = {inner_product}")
print(f"Cosine of the angle: {cos_theta:.2f}")
print(f"Angle θ: {theta_degrees:.2f} degrees")

# Plot the vectors
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

# Draw vectors
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')

# Add grid and labels
ax.grid(True, linestyle='--', linewidth=0.5)
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('2D Vectors and Inner Product')
ax.legend()

plt.show()
````
